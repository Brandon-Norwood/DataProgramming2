## Number of counties in each state

Use a dictionary to create a histogram of the state values from the "uscounties.csv" file. Neatly print the results.

Expected output:
<pre>
Number of counties (by state)
-----------------------------
Alabama : 67
Alaska : 29
Arizona : 15
...
West Virginia : 55
Wisconsin : 72
Wyoming : 23
</pre>

```python
counties_state = {}

fHandle = open(r"uscounties1.csv")

next(fHandle)

for line in fHandle:
    line = line.strip()
    fields = line.split(",")
    state = fields[5]
    
    counties_state[state] = counties_state.get(state, 0) +1
        
for key, value in sorted(counties_state.items()):
    print(f"{key}: {value}")
        
    

```

## Number of counties by designation

Use a dictionary to create a histogram of the county designations from the "uscounties.csv" file. The designation of a county is the last word in the county's *full name*.

Neatly print the results.

Reference: <a href="https://stackoverflow.com/questions/1823058/how-to-print-number-with-commas-as-thousands-separators#:~:text=f%27%7Bvalue%3A%2C%7D%27%20%20%23%20For%20Python%20%E2%89%A53.6">How to print a number with commas as thousands separators</a>. You're using *Python ≥3.6*.

Expected output:
<pre>
Number of counties (by designation)
-----------------------------------
Area : 10
Borough : 17
City : 41
Columbia : 1
County : 3,007
Municipality : 2
Parish : 64
</pre>

Yes, I know "Columbia" is not a designation. Just <a href="https://www.youtube.com/watch?v=L0MK7qz13bU">let it go</a>.

```python
counties_state = {}

fHandle = open(r"uscounties1.csv")

next(fHandle)

for line in fHandle:
    line = line.strip()
    fields = line.split(",")
    state = fields[5]
    countyName = fields[2]
    designation = countyName.split(" ")
    designation = designation[-1]
    
    counties_state[designation] = counties_state.get(designation, 0) +1

for key, value in sorted(counties_state.items()):
    print(f"{key}: {value}")
```

## Number of counties by population

Use a dictionary to create a histogram of the counties in the U.S., grouped by population. Specifically, group counties based on increments of 100,000.

For example, the first group would be states with a population less than 100,000. The second group would be states with a population 100,000 to 199,999. And so on. *Hint: You'll be dividing(?) the population of each county by 100,000.*

Neatly print the results.

Reference: <a href="https://stackoverflow.com/questions/1823058/how-to-print-number-with-commas-as-thousands-separators#:~:text=f%27%7Bvalue%3A%2C%7D%27%20%20%23%20For%20Python%20%E2%89%A53.6">How to print a number with commas as thousands separators</a>. You're using *Python ≥3.6*.

Expected output:
<pre>
Number of counties (by population)
----------------------------------
0 - 99,999 : 2,546
100,000 - 199,999 : 264
200,000 - 299,999 : 101
...
4,600,000 - 4,699,999 : 1
5,100,000 - 5,199,999 : 1
10,000,000 - 10,099,999 : 1
</pre>

```python
county_population = {}

fHandle = open(r"uscounties1.csv")
next(fHandle)

for line in fHandle:
    line = line.strip()
    line = line.replace('"',"")
    fields = line.split(",")
    population = int(fields[8])/100000
    population = str(population)
    population = population.split(".")
    population = population[0]
    
    county_population[population] = county_population.get(population, 0) +1
    
for key, value in sorted(county_population.items()):
    key = int(key) * 100000
    print(f"{key:,} - {(key+99999):,}: {value}")
    
    
```
