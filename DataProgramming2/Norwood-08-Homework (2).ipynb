## Computing course numbers

Build a list of the unique course numbers found in the `computing_courses.txt` file. Neatly print the sorted list along with the number of unique prefixes.

*Note: A "course number" is the numeric portion of the course identifier. For example, "**2431**" is the* course number *for the course "IT 2431". Couirse numbers may include a trailing letter such as "**5090G**".*

Sample output:
<pre>1130
1230
1231
...
7891
7895
7999
83 prefixes
</pre>

```python
fHandle = open(r'Untitled Folder/computing_courses.txt')

num = []

for line in fHandle:   
        
    index = line.find(' ')
    line = line[index+1:]
    
    index = line.find(',')
    courseNum = line[:index]
    num.append(courseNum)

for itr in num:
    print(itr)

print(len(num), "prefixes")
   
```

## Z Words in course descriptions

Build a list of unique words starting with the letter "Z" from the `course_descriptions.txt` file. Ignore differences in case (e.g. 'make' and 'Make' are the same) and store the lowercase form of the words. Neatly print the sorted list along with the number of unique words.

*Note: Be sure to <a href="https://stackoverflow.com/questions/265960/best-way-to-strip-punctuation-from-a-string#266162">filter out the punctuation</a> from the words.*

Sample output:
<pre>Z
Zealand
Zooarchaeology
...
zooarchaeology
ztransforms
zygote
12 total words
</pre>

```python
#words = []
zWords = []
fHandle = open(r'Untitled Folder/course_descriptions.txt')

for line in fHandle:
    line = line.strip()
    line = line.split(" ")
    
    for word in line:
        if word.startswith("z") or word.startswith("Z"):
            zWords.append(word)

zWords = list(set(zWords))
for word in zWords:
    print(word)
print(len(zWords), "total words")
```

## Words longer than fourteen characters

Build a unique list of words with more than 14 characters from the `hound.txt` file. Neatly print the sorted list along with the number of words.

*Bonus: Strip off trailing punctuation*

Sample output:
<pre>anthropological
characteristics
dolichocephalic
extraordinarily
personification
supernaturalists
6 total words
</pre>

```python
fHandle = open(r'Untitled Folder/hound.txt')

bigWords = []

for line in fHandle:
    line = line.strip()
    line = line.split(" ")
    
    for word in line:
        if len(word) > 14:
            bigWords.append(word)

bigWords = list(set(bigWords))            
for word in bigWords:
    print(word)

print(len(bigWords), "total words")
    




```

## 100 Random numbers

Build a list of 100 <a href="https://www.w3schools.com/python/ref_random_randint.asp">random numbers</a> between 1 and 999. Print the lowest and highest values from the list along with the average of all numbers.

Sample output:
<pre>Lowest: 4
Highest: 995
Average: 496.14
</pre>

```python
import random

numList = []

for num in range(1,101):
    num = random.randint(1,999)
    numList.append(num)

highestNum = max(numList)
lowestNum = min(numList)
averageNum = sum(numList)/len(numList)

print("Lowest:", lowestNum)
print("Highest:", highestNum)
print("Average:", averageNum)
```
